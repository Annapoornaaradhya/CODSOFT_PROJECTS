```python
#TO-DO LIST
#TASK 1
import tkinter as tk

def add_task():
    task = entry.get()
    if task:
        listbox.insert(tk.END, task)
        entry.delete(0, tk.END)

def delete_task():
    selected_task = listbox.curselection()
    if selected_task:
        listbox.delete(selected_task)


root = tk.Tk()
root.title("To-Do List")


entry = tk.Entry(root, width=30)
entry.grid(row=0, column=0, padx=10, pady=10)


add_button = tk.Button(root, text="Add Task", command=add_task)
add_button.grid(row=0, column=1, padx=10, pady=10)


listbox = tk.Listbox(root, selectmode=tk.SINGLE, width=30)
listbox.grid(row=1, column=0, columnspan=2, padx=10, pady=10)

delete_button = tk.Button(root, text="Delete Task", command=delete_task)
delete_button.grid(row=2, column=0, columnspan=2, pady=10)


root.mainloop()
```

```python
#SIMPLE CALCULATOR 
#TASK 2
import tkinter as tk

def on_click(button_text):
    current_text = result_var.get()
    if button_text == "=":
        try:
            result_var.set(eval(current_text))
        except Exception as e:
            result_var.set("Error")
    elif button_text == "C":
        result_var.set("")
    else:
        result_var.set(current_text + button_text)


root = tk.Tk()
root.title("Simple Calculator")


result_var = tk.StringVar()


result_entry = tk.Entry(root, textvariable=result_var, justify="right", font=("Helvetica", 16))
result_entry.grid(row=0, column=0, columnspan=4, padx=10, pady=10, ipadx=8, ipady=8)


buttons = [
    "7", "8", "9", "/",
    "4", "5", "6", "*",
    "1", "2", "3", "-",
    "0", ".", "=", "+",
    "C"
]


row_val = 1
col_val = 0
for button_text in buttons:
    tk.Button(root, text=button_text, width=5, height=2, command=lambda b=button_text: on_click(b)).grid(row=row_val, column=col_val)
    col_val += 1
    if col_val > 3:
        col_val = 0
        row_val += 1


root.mainloop()
```

```python
#PASSWORD GENERATOR
#TASK 3

import random
import string

def generate_password(name, length=12):
    chars = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(chars) for _ in range(length))
    return f"{name}'s Password: {password}"


try:
    name = input("Enter your name: ")
    length = int(input("Enter the password length: "))

    if length <= 0:
        print("Invalid length. Please enter a positive number.")
    else:
        password = generate_password(name, length)
        print("Generated Password:", password)
except ValueError:
    print("Invalid input. Please enter a valid number.")
```
